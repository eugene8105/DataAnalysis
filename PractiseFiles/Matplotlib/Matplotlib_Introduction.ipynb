<h2>Introduction to Matplotlib</h2>
<h4>Matplotlib Part 1</h4>

```python
import matplotlib.pyplot as plt
```

<h4>Will allowed you see plots inside the jupyter notebook</h4>

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional
plt.plot(x,y)
#if not in jupyter notebook use
#plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# creating multiple plots on the same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object orianted method
fig = plt.figure()
# fig.add_axes([left of the axes,button,width,height])
axes = fig.add_axes([0.1,0.1,0.8,0.8])
# plot on the axes
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Lable')
axes.set_title('Titel')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('smaller plot')
```

<h4>Matplotlib Part 2</h4>

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
# fixing overlapping
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Axes 0')
axes[1].plot(y,x)
axes[1].set_title('Axes 1')
```

<h2>Figure Size and DPI</h2>

```python
fig = plt.figure(figsize=(3,2))
```

```python
fig,ax = plt.subplots(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

<h1>Matplotlib Part 3</h1>

<h2>Plot Appearance</h2>

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# the alpha - how transparent the line is.
# linewidth can be short lw, linestyle - ls
# linestyle can be set up with '--',':','steps',...
# marker - will mark points
ax.plot(x,y,color='green',lw=3,alpha=0.5,ls='-',marker='o',markersize=20,
        markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```

```python
# lets look on x - it's array with 11 points
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# the alpha
ax.plot(x,y,color='green',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
